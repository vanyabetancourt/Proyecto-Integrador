```python
# ============================================================
# 2. IMPORTAR LIBRERÍAS
# ============================================================

import os
import cv2
import zipfile
import shutil
import numpy as np
import pandas as pd
import mediapipe as mp
import matplotlib.pyplot as plt

from google.colab import drive, files

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
```

```python
# ============================================================
# 4. CONFIGURACIÓN PARA SEÑAS DINÁMICAS
# ============================================================

# Ruta donde está tu ZIP de videos dinámicos en Drive
# Ajusta el nombre si tu archivo se llama diferente
DYNAMIC_ZIP_PATH = "/content/drive/MyDrive/Colab Notebooks/MNA/Proyecto Integrador/Proyecto Integrador.zip"

# Carpeta temporal donde se descomprimirán los videos
DYNAMIC_EXTRACT_PATH = "/content/lsm_dynamic_videos"

# Número de frames por video
SEQUENCE_LENGTH = 40

# MediaPipe Hands detecta 21 puntos por mano
NUM_LANDMARKS = 21

# 21 puntos x 3 coordenadas x 2 manos = 126 features
NUM_FEATURES = NUM_LANDMARKS * 3 * 2

VIDEO_EXTENSIONS = (".mp4", ".avi", ".mov", ".mkv")

print("Configuración dinámica lista.")
```

```python
# ============================================================
# 5. DESCOMPRIMIR VIDEOS DINÁMICOS
# ============================================================

os.makedirs(DYNAMIC_EXTRACT_PATH, exist_ok=True)

if os.path.exists(DYNAMIC_ZIP_PATH):
    with zipfile.ZipFile(DYNAMIC_ZIP_PATH, 'r') as zip_ref:
        zip_ref.extractall(DYNAMIC_EXTRACT_PATH)

    print("Videos dinámicos descomprimidos en:", DYNAMIC_EXTRACT_PATH)
else:
    print("No se encontró el ZIP dinámico. Revisa esta ruta:")
    print(DYNAMIC_ZIP_PATH)
```

```python
# ============================================================
# 6. LOCALIZAR CARPETA CON VIDEOS MP4
# ============================================================

DYNAMIC_DATASET_PATH = None

for root, dirs, files_list in os.walk(DYNAMIC_EXTRACT_PATH):
    mp4_files = [f for f in files_list if f.lower().endswith(VIDEO_EXTENSIONS)]

    if mp4_files:
        DYNAMIC_DATASET_PATH = root
        print("Carpeta con videos encontrada:", DYNAMIC_DATASET_PATH)
        print("Cantidad de videos:", len(mp4_files))
        print("Ejemplos:", mp4_files[:10])
        break

if DYNAMIC_DATASET_PATH is None:
    print("No se encontraron videos. Revisa el ZIP o la ruta.")
```

```python
# ============================================================
# 7. CONFIGURAR MEDIAPIPE PARA VIDEOS
# ============================================================

mp_hands = mp.solutions.hands

dynamic_hands = mp_hands.Hands(
    static_image_mode=False,
    max_num_hands=2,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

print("MediaPipe para video configurado.")
```

```python
# ============================================================
# 8. FUNCIONES PARA PROCESAR VIDEOS DINÁMICOS
# ============================================================

def get_label_from_filename(filename):
    """
    Extrae la clase desde el nombre del archivo.

    Ejemplo:
    S1-J-frontal-1.mp4 -> J
    S1-K-perfil-2.mp4  -> K
    """

    name_without_ext = os.path.splitext(filename)[0]
    parts = name_without_ext.split("-")

    if len(parts) < 2:
        return None

    return parts[1]


def extract_dynamic_hand_landmarks(frame):
    """
    Extrae landmarks de ambas manos desde un frame de video.

    Salida:
    Vector de 126 valores:
    21 landmarks x 3 coordenadas x 2 manos.
    """

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    results = dynamic_hands.process(frame_rgb)

    left_hand = np.zeros(NUM_LANDMARKS * 3)
    right_hand = np.zeros(NUM_LANDMARKS * 3)

    if results.multi_hand_landmarks and results.multi_handedness:
        for hand_landmarks, handedness in zip(
            results.multi_hand_landmarks,
            results.multi_handedness
        ):
            hand_label = handedness.classification[0].label

            landmarks = []

            for lm in hand_landmarks.landmark:
                landmarks.extend([lm.x, lm.y, lm.z])

            landmarks = np.array(landmarks)

            if hand_label == "Left":
                left_hand = landmarks
            elif hand_label == "Right":
                right_hand = landmarks

    return np.concatenate([left_hand, right_hand])


def process_video(video_path, sequence_length=SEQUENCE_LENGTH):
    """
    Convierte un video en una secuencia de landmarks.

    Entrada:
    video_path: ruta del video.

    Salida:
    Array de forma:
    40 frames x 126 features.
    """

    cap = cv2.VideoCapture(video_path)

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames == 0:
        cap.release()
        return None

    frame_indices = np.linspace(0, total_frames - 1, sequence_length).astype(int)

    sequence = []

    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)

        ret, frame = cap.read()

        if not ret:
            sequence.append(np.zeros(NUM_FEATURES))
            continue

        landmarks = extract_dynamic_hand_landmarks(frame)

        sequence.append(landmarks)

    cap.release()

    return np.array(sequence)
```

```python
# ============================================================
# 9. PROCESAR DATASET DINÁMICO
# ============================================================

X_dynamic = []
y_dynamic = []

videos = [
    archivo for archivo in os.listdir(DYNAMIC_DATASET_PATH)
    if archivo.lower().endswith(VIDEO_EXTENSIONS)
]

for idx, file_name in enumerate(videos, start=1):

    label = get_label_from_filename(file_name)

    if label is None:
        print("Archivo ignorado:", file_name)
        continue

    video_path = os.path.join(DYNAMIC_DATASET_PATH, file_name)

    sequence = process_video(video_path)

    if sequence is not None:
        X_dynamic.append(sequence)
        y_dynamic.append(label)
        print(f"{idx}/{len(videos)} procesado: {file_name} → {label}")
    else:
        print(f"{idx}/{len(videos)} no se pudo procesar: {file_name}")

X_dynamic = np.array(X_dynamic)
y_dynamic = np.array(y_dynamic)

print("Dataset dinámico procesado.")
print("X_dynamic:", X_dynamic.shape)
print("y_dynamic:", y_dynamic.shape)
print("Clases dinámicas:", sorted(set(y_dynamic)))
```

```python
# ============================================================
# 10. GUARDAR LANDMARKS DINÁMICOS
# ============================================================

DYNAMIC_SAVE_PATH = os.path.join(BASE_SAVE_PATH, "dynamic_model")

os.makedirs(DYNAMIC_SAVE_PATH, exist_ok=True)

np.save(os.path.join(DYNAMIC_SAVE_PATH, "X_dynamic_landmarks.npy"), X_dynamic)
np.save(os.path.join(DYNAMIC_SAVE_PATH, "y_dynamic_labels.npy"), y_dynamic)

print("Landmarks dinámicos guardados.")
```

```python
# ============================================================
# 11. ENTRENAR MODELO DINÁMICO BiLSTM
# ============================================================

dynamic_label_encoder = LabelEncoder()

y_dynamic_encoded = dynamic_label_encoder.fit_transform(y_dynamic)
y_dynamic_cat = to_categorical(y_dynamic_encoded)

X_dynamic_train, X_dynamic_temp, y_dynamic_train, y_dynamic_temp, y_dynamic_encoded_train, y_dynamic_encoded_temp = train_test_split(
    X_dynamic,
    y_dynamic_cat,
    y_dynamic_encoded,
    test_size=0.30,
    random_state=42,
    stratify=y_dynamic_encoded
)

X_dynamic_val, X_dynamic_test, y_dynamic_val, y_dynamic_test = train_test_split(
    X_dynamic_temp,
    y_dynamic_temp,
    test_size=0.50,
    random_state=42,
    stratify=y_dynamic_encoded_temp
)

num_dynamic_classes = len(dynamic_label_encoder.classes_)

dynamic_model = Sequential([
    Bidirectional(
        LSTM(128, return_sequences=True),
        input_shape=(SEQUENCE_LENGTH, NUM_FEATURES)
    ),
    BatchNormalization(),
    Dropout(0.4),

    Bidirectional(
        LSTM(64, return_sequences=False)
    ),
    BatchNormalization(),
    Dropout(0.4),

    Dense(128, activation="relu"),
    Dropout(0.3),

    Dense(64, activation="relu"),
    Dropout(0.3),

    Dense(num_dynamic_classes, activation="softmax")
])

dynamic_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

dynamic_model.summary()
```

```python
# ============================================================
# 12. ENTRENAR Y GUARDAR MODELO DINÁMICO
# ============================================================

dynamic_checkpoint = ModelCheckpoint(
    os.path.join(DYNAMIC_SAVE_PATH, "modelo_lsm_dinamico_bilstm.keras"),
    monitor="val_accuracy",
    save_best_only=True,
    mode="max"
)

dynamic_early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=12,
    restore_best_weights=True
)

dynamic_reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=5,
    min_lr=1e-6
)

dynamic_history = dynamic_model.fit(
    X_dynamic_train,
    y_dynamic_train,
    validation_data=(X_dynamic_val, y_dynamic_val),
    epochs=100,
    batch_size=16,
    callbacks=[dynamic_checkpoint, dynamic_early_stopping, dynamic_reduce_lr]
)

np.save(
    os.path.join(DYNAMIC_SAVE_PATH, "clases_lsm_dinamicas.npy"),
    dynamic_label_encoder.classes_
)

print("Modelo dinámico guardado.")
```

```python
# ============================================================
# FUNCIÓN GENERAL DE EVALUACIÓN
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

def evaluate_model_visual(
    model,
    X_test,
    y_test,
    class_names,
    model_name="Modelo"
):
    """
    1. Cards con Accuracy, Precision, Recall y F1-score.
    2. Matriz de confusión absoluta.
    3. Matriz de confusión porcentual.
    4. Tabla visual de precision, recall y f1-score por clase.
    5. Tabla de predicciones.
    6. Tabla de errores.
    """

    # --------------------------------------------------------
    # 1. Obtener predicciones
    # --------------------------------------------------------

    y_pred_probs = model.predict(X_test)

    # Clase predicha
    y_pred = np.argmax(y_pred_probs, axis=1)

    # Clase real
    # Si y_test viene en one-hot, se convierte a índice
    if len(y_test.shape) > 1:
        y_true = np.argmax(y_test, axis=1)
    else:
        y_true = y_test

    # --------------------------------------------------------
    # 2. Métricas generales
    # --------------------------------------------------------

    accuracy = accuracy_score(y_true, y_pred) * 100
    precision = precision_score(y_true, y_pred, average="weighted", zero_division=0) * 100
    recall = recall_score(y_true, y_pred, average="weighted", zero_division=0) * 100
    f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0) * 100

    metrics_df = pd.DataFrame({
        "Metric": [
            "Accuracy",
            "Precision Weighted",
            "Recall Weighted",
            "F1-score Weighted"
        ],
        "Value": [
            f"{accuracy:.2f}%",
            f"{precision:.2f}%",
            f"{recall:.2f}%",
            f"{f1:.2f}%"
        ]
    })

    print("===================================================")
    print(f" RESULTADOS GENERALES - {model_name}")
    print("===================================================")
    display(metrics_df)

    # --------------------------------------------------------
    # 3. Matriz de confusión absoluta
    # --------------------------------------------------------

    cm = confusion_matrix(y_true, y_pred)

    fig, ax = plt.subplots(figsize=(8, 6))

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=class_names
    )

    disp.plot(
        cmap="Blues",
        values_format="d",
        ax=ax,
        colorbar=True
    )

    plt.title(f"Matriz de Confusión - {model_name}", fontsize=14)
    plt.xlabel("Clase predicha", fontsize=12)
    plt.ylabel("Clase real", fontsize=12)
    plt.grid(False)
    plt.show()

    # --------------------------------------------------------
    # 4. Matriz de confusión porcentual
    # --------------------------------------------------------

    cm_percent = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] * 100
    cm_percent = np.nan_to_num(cm_percent)

    fig, ax = plt.subplots(figsize=(8, 6))

    disp_percent = ConfusionMatrixDisplay(
        confusion_matrix=cm_percent,
        display_labels=class_names
    )

    disp_percent.plot(
        cmap="Blues",
        values_format=".1f",
        ax=ax,
        colorbar=True
    )

    plt.title(f"Matriz de Confusión (%) - {model_name}", fontsize=14)
    plt.xlabel("Clase predicha", fontsize=12)
    plt.ylabel("Clase real", fontsize=12)
    plt.grid(False)
    plt.show()

    # --------------------------------------------------------
    # 5. Reporte de clasificación por clase
    # --------------------------------------------------------

    report = classification_report(
        y_true,
        y_pred,
        target_names=class_names,
        output_dict=True,
        zero_division=0
    )

    report_df = pd.DataFrame(report).transpose()

    # Convertir métricas a porcentaje
    for col in ["precision", "recall", "f1-score"]:
        if col in report_df.columns:
            report_df[col] = report_df[col] * 100

    report_df = report_df.round(2)

    print("===================================================")
    print(f" MÉTRICAS POR CLASE - {model_name}")
    print("===================================================")

    display(report_df)

    # --------------------------------------------------------
    # 6. Gráfica de métricas por clase
    # --------------------------------------------------------

    class_report_df = report_df.loc[class_names, ["precision", "recall", "f1-score"]]

    ax = class_report_df.plot(
        kind="bar",
        figsize=(10, 5)
    )

    plt.title(f"Precision, Recall y F1-score por clase - {model_name}", fontsize=14)
    plt.xlabel("Clase", fontsize=12)
    plt.ylabel("Porcentaje (%)", fontsize=12)
    plt.ylim(0, 105)
    plt.xticks(rotation=0)
    plt.grid(axis="y", alpha=0.3)
    plt.legend(title="Métrica")
    plt.show()

    # --------------------------------------------------------
    # 7. Tabla de predicciones
    # --------------------------------------------------------

    results_df = pd.DataFrame({
        "True Label": np.array(class_names)[y_true],
        "Predicted Label": np.array(class_names)[y_pred],
        "Confidence (%)": np.max(y_pred_probs, axis=1) * 100
    })

    results_df["Correct"] = results_df["True Label"] == results_df["Predicted Label"]
    results_df["Confidence (%)"] = results_df["Confidence (%)"].round(2)

    print("===================================================")
    print(f" TABLA DE PREDICCIONES - {model_name}")
    print("===================================================")

    display(results_df)

    # --------------------------------------------------------
    # 8. Tabla de errores
    # --------------------------------------------------------

    errors_df = results_df[results_df["Correct"] == False]

    print("===================================================")
    print(f" ANÁLISIS DE ERRORES - {model_name}")
    print("===================================================")
    print("Total de muestras:", len(results_df))
    print("Predicciones correctas:", results_df["Correct"].sum())
    print("Predicciones incorrectas:", len(errors_df))
    print("Error rate:", round((len(errors_df) / len(results_df)) * 100, 2), "%")

    display(errors_df)

    return {
        "metrics_summary": metrics_df,
        "classification_report": report_df,
        "predictions": results_df,
        "errors": errors_df,
        "confusion_matrix": cm,
        "confusion_matrix_percent": cm_percent
    }
```

```python
# ============================================================
# CREAR DATAFRAME DE IMÁGENES ESTÁTICAS
# TOMANDO LA CLASE DESDE EL NOMBRE DEL ARCHIVO
# ============================================================

import os
import pandas as pd

STATIC_DATASET_ROOT = "/content/msl_abc_v2"

IMAGE_EXTENSIONS = (".jpg", ".jpeg", ".png", ".webp")

static_samples = []

def get_static_label_from_filename(filename):
    """
    Extrae la letra desde el nombre del archivo.

    Ejemplos:
    S6-D-4-140.jpg  -> D
    S5-W-4-32.jpg   -> W
    S1-A-4-113.jpg  -> A

    Estructura esperada:
    S#-LETRA-#-#.jpg
    """

    # Quitar extensión
    name_without_ext = os.path.splitext(filename)[0]

    # Separar por guion
    parts = name_without_ext.split("-")

    # Validar que tenga al menos dos partes
    if len(parts) < 2:
        return None

    # La letra está en la segunda posición
    label = parts[1]

    return label


# Recorrer todo el dataset
for root, dirs, files in os.walk(STATIC_DATASET_ROOT):

    for file_name in files:

        if file_name.lower().endswith(IMAGE_EXTENSIONS):

            label = get_static_label_from_filename(file_name)

            if label is not None:
                static_samples.append({
                    "image_path": os.path.join(root, file_name),
                    "label": label,
                    "file_name": file_name,
                    "folder": root
                })

static_df = pd.DataFrame(static_samples)

print("Total de imágenes encontradas:", len(static_df))

print("\nConteo por clase:")
print(static_df["label"].value_counts().sort_index())

print("\nNúmero de clases:", static_df["label"].nunique())

display(static_df.head())
```

```python
# ============================================================
# CONFIGURAR MEDIAPIPE PARA IMÁGENES ESTÁTICAS
# ============================================================

import cv2
import numpy as np
import mediapipe as mp

# MediaPipe Hands detecta 21 puntos por mano
NUM_LANDMARKS = 21

# 21 puntos x 3 coordenadas x 2 manos = 126 features
NUM_FEATURES_STATIC = NUM_LANDMARKS * 3 * 2

# Inicializar MediaPipe Hands
mp_hands = mp.solutions.hands

# Para imágenes usamos static_image_mode=True
static_hands = mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5
)

print("MediaPipe para imágenes estáticas configurado.")
print("Features por imagen:", NUM_FEATURES_STATIC)
```

```python
# ============================================================
# FUNCIÓN PARA EXTRAER LANDMARKS DE UNA IMAGEN
# ============================================================

def extract_static_image_landmarks(image_path):
    """
    Convierte una imagen de una seña estática en un vector de landmarks.

    Entrada:
    - image_path: ruta de la imagen.

    Salida:
    - vector de 126 valores:
      21 puntos x 3 coordenadas x 2 manos.

    Si no detecta una mano, rellena con ceros.
    """

    # Leer imagen
    image = cv2.imread(image_path)

    # Validar que la imagen se leyó correctamente
    if image is None:
        return None

    # Convertir de BGR a RGB porque MediaPipe usa RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Procesar imagen con MediaPipe
    results = static_hands.process(image_rgb)

    # Vectores vacíos para cada mano
    left_hand = np.zeros(NUM_LANDMARKS * 3)
    right_hand = np.zeros(NUM_LANDMARKS * 3)

    # Si detecta manos, extraer coordenadas
    if results.multi_hand_landmarks and results.multi_handedness:
        for hand_landmarks, handedness in zip(
            results.multi_hand_landmarks,
            results.multi_handedness
        ):
            hand_label = handedness.classification[0].label

            landmarks = []

            for lm in hand_landmarks.landmark:
                landmarks.extend([lm.x, lm.y, lm.z])

            landmarks = np.array(landmarks)

            if hand_label == "Left":
                left_hand = landmarks
            elif hand_label == "Right":
                right_hand = landmarks

    # Unir ambas manos
    return np.concatenate([left_hand, right_hand])
```

```python
# ============================================================
# PROCESAR DATASET ESTÁTICO
# ============================================================

X_static = []
y_static = []

for idx, row in static_df.iterrows():

    image_path = row["image_path"]
    label = row["label"]

    landmarks = extract_static_image_landmarks(image_path)

    if landmarks is not None:
        X_static.append(landmarks)
        y_static.append(label)

    if idx % 1000 == 0:
        print(f"Procesadas {idx}/{len(static_df)} imágenes")

X_static = np.array(X_static)
y_static = np.array(y_static)

print("Procesamiento terminado.")
print("X_static:", X_static.shape)
print("y_static:", y_static.shape)
print("Clases detectadas:", sorted(set(y_static)))
```

```python
# ============================================================
# GUARDAR LANDMARKS ESTÁTICOS EN DRIVE
# ============================================================

import os

STATIC_SAVE_PATH = "/content/drive/MyDrive/LSM_Hibrido/static_model"

os.makedirs(STATIC_SAVE_PATH, exist_ok=True)

np.save(os.path.join(STATIC_SAVE_PATH, "X_static_landmarks.npy"), X_static)
np.save(os.path.join(STATIC_SAVE_PATH, "y_static_labels.npy"), y_static)

print("Landmarks estáticos guardados en:")
print(STATIC_SAVE_PATH)
```

```python
# ============================================================
# CODIFICAR CLASES Y SEPARAR DATASET
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Convertir letras a números
static_label_encoder = LabelEncoder()
y_static_encoded = static_label_encoder.fit_transform(y_static)

# One-hot encoding
y_static_cat = to_categorical(y_static_encoded)

print("Clases codificadas:")
for idx, clase in enumerate(static_label_encoder.classes_):
    print(idx, "→", clase)

# 70% train, 30% temporal
X_static_train, X_static_temp, y_static_train, y_static_temp, y_static_encoded_train, y_static_encoded_temp = train_test_split(
    X_static,
    y_static_cat,
    y_static_encoded,
    test_size=0.30,
    random_state=42,
    stratify=y_static_encoded
)

# 15% validation, 15% test
X_static_val, X_static_test, y_static_val, y_static_test = train_test_split(
    X_static_temp,
    y_static_temp,
    test_size=0.50,
    random_state=42,
    stratify=y_static_encoded_temp
)

print("Train:", X_static_train.shape)
print("Validation:", X_static_val.shape)
print("Test:", X_static_test.shape)
```

```python
# ============================================================
# CREAR MODELO ESTÁTICO MLP
# ============================================================

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

num_static_classes = len(static_label_encoder.classes_)

static_model = Sequential([
    Dense(256, activation="relu", input_shape=(NUM_FEATURES_STATIC,)),
    BatchNormalization(),
    Dropout(0.4),

    Dense(128, activation="relu"),
    BatchNormalization(),
    Dropout(0.3),

    Dense(64, activation="relu"),
    Dropout(0.3),

    Dense(num_static_classes, activation="softmax")
])

static_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

static_model.summary()
```

```python
# ============================================================
# ENTRENAR MODELO ESTÁTICO
# ============================================================

static_checkpoint = ModelCheckpoint(
    os.path.join(STATIC_SAVE_PATH, "modelo_lsm_estatico_mlp.keras"),
    monitor="val_accuracy",
    save_best_only=True,
    mode="max"
)

static_early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

static_reduce_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=4,
    min_lr=1e-6
)

static_history = static_model.fit(
    X_static_train,
    y_static_train,
    validation_data=(X_static_val, y_static_val),
    epochs=80,
    batch_size=32,
    callbacks=[static_checkpoint, static_early_stopping, static_reduce_lr]
)

# Guardar clases
np.save(
    os.path.join(STATIC_SAVE_PATH, "clases_lsm_estaticas.npy"),
    static_label_encoder.classes_
)

print("Modelo estático y clases guardados.")
```

```python
# ============================================================
# EVALUACIÓN VISUAL DEL MODELO ESTÁTICO
# ============================================================

from IPython.display import display
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

import pandas as pd
import matplotlib.pyplot as plt

# Predicciones
y_pred_probs = static_model.predict(X_static_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_static_test, axis=1)

class_names = static_label_encoder.classes_

# Métricas generales
accuracy = accuracy_score(y_true, y_pred) * 100
precision = precision_score(y_true, y_pred, average="weighted", zero_division=0) * 100
recall = recall_score(y_true, y_pred, average="weighted", zero_division=0) * 100
f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0) * 100

metrics_df = pd.DataFrame({
    "Métrica": ["Accuracy", "Precision", "Recall", "F1-score"],
    "Resultado": [
        f"{accuracy:.2f}%",
        f"{precision:.2f}%",
        f"{recall:.2f}%",
        f"{f1:.2f}%"
    ]
})

print("RESULTADOS GENERALES - MODELO ESTÁTICO")
display(metrics_df)

# Matriz de confusión absoluta
cm = confusion_matrix(y_true, y_pred)

fig, ax = plt.subplots(figsize=(12, 10))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=class_names
)
disp.plot(cmap="Blues", values_format="d", ax=ax)
plt.title("Matriz de Confusión - Modelo Estático")
plt.xlabel("Predicción")
plt.ylabel("Clase real")
plt.xticks(rotation=90)
plt.grid(False)
plt.show()

# Matriz de confusión porcentual
cm_percent = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] * 100
cm_percent = np.nan_to_num(cm_percent)

fig, ax = plt.subplots(figsize=(12, 10))
disp_percent = ConfusionMatrixDisplay(
    confusion_matrix=cm_percent,
    display_labels=class_names
)
disp_percent.plot(cmap="Blues", values_format=".1f", ax=ax)
plt.title("Matriz de Confusión (%) - Modelo Estático")
plt.xlabel("Predicción")
plt.ylabel("Clase real")
plt.xticks(rotation=90)
plt.grid(False)
plt.show()

# Classification report
report = classification_report(
    y_true,
    y_pred,
    target_names=class_names,
    output_dict=True,
    zero_division=0
)

report_df = pd.DataFrame(report).transpose()

for col in ["precision", "recall", "f1-score"]:
    report_df[col] = report_df[col] * 100

report_df = report_df.round(2)

print("MÉTRICAS POR CLASE - MODELO ESTÁTICO")
display(report_df)

# Gráfica de métricas por clase
class_metrics_df = report_df.loc[class_names, ["precision", "recall", "f1-score"]]

class_metrics_df.plot(kind="bar", figsize=(14, 6))
plt.title("Precision, Recall y F1-score por clase - Modelo Estático")
plt.xlabel("Clase")
plt.ylabel("Porcentaje (%)")
plt.ylim(0, 105)
plt.xticks(rotation=90)
plt.grid(axis="y", alpha=0.3)
plt.show()
```
